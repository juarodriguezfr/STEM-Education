````python
class IrisDataIterator:
    def __init__(self, file_path):
        self.data = pd.read_csv(file_path)
        self.index = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.index < len(self.data):
            data_chunk = self.data.iloc[self.index:self.index + 2]
            self.index += 2
            return data_chunk
        else:
            raise StopIteration("error")
````

Probamos la clase:

````python
if __name__ == "__main__":
    iris_iterator = IrisDataIterator(file_path)
    try:
        while True:
            data_chunk = next(iris_iterator)
            print(data_chunk)
    except StopIteration as e:
        print(e)
````
